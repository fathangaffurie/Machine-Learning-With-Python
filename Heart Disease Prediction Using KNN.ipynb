```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier as KNNCf
from jupyterthemes import jtplot
jtplot.style(theme='onedork')
```

```python
datas = data.copy()
#feature_df = datas[["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal"]]
feature_df = datas[["age","trestbps","chol","thalach","oldpeak"]]
x = np.asarray(feature_df)
y = np.asarray(data["target"].astype('int'))
```

```python
print(feature_df.shape)
```

```python
knn = KNNCf(n_neighbors=2, weights='uniform',algorithm='auto',metric='euclidean')
#fitting model
knn.fit(x,y)
```

```python
newinput = [[63,145,233,150,2.3],[60,130,253,144,1.4],[63,145,233,150,2.3]]
```

```python
pred = knn.predict(newinput)
for i in pred:
    if i == 1:
        print("positif")
    else:
        print("negatif")
```
